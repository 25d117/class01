```python
# prompt: 선택구조프로그래밍을 이용해서 로봇 청소기 프로그래밍을 만들어줘

class RobotVacuum:
    def __init__(self, battery_level=100, location=(0, 0)):
        self.battery_level = battery_level
        self.location = location
        self.is_cleaning = False
        self.is_charging = False
        self.is_stuck = False

    def display_status(self):
        print(f"배터리 잔량: {self.battery_level}%")
        print(f"현재 위치: {self.location}")
        print(f"청소 중: {self.is_cleaning}")
        print(f"충전 중: {self.is_charging}")
        print(f"낑김 상태: {self.is_stuck}")
        print("-" * 20)

    def start_cleaning(self):
        if self.battery_level > 20 and not self.is_charging and not self.is_stuck:
            print("청소를 시작합니다.")
            self.is_cleaning = True
        elif self.battery_level <= 20:
            print("배터리가 부족하여 청소를 시작할 수 없습니다. 충전이 필요합니다.")
            self.move_to_charging_station() # 충전 스테이션으로 이동하는 로직 추가
        elif self.is_charging:
            print("현재 충전 중입니다.")
        elif self.is_stuck:
            print("로봇이 낑겨 있습니다. 도움을 기다립니다.")

    def stop_cleaning(self):
        if self.is_cleaning:
            print("청소를 중단합니다.")
            self.is_cleaning = False
        else:
            print("현재 청소 중이 아닙니다.")

    def move(self, direction):
        if self.is_cleaning and not self.is_stuck:
            print(f"{direction} 방향으로 이동합니다.")
            # 실제 위치 이동 로직 (예: self.location 업데이트)
            if direction == "앞":
                self.location = (self.location[0] + 1, self.location[1])
            elif direction == "뒤":
                self.location = (self.location[0] - 1, self.location[1])
            elif direction == "왼쪽":
                self.location = (self.location[0], self.location[1] - 1)
            elif direction == "오른쪽":
                self.location = (self.location[0], self.location[1] + 1)
            # 배터리 소모
            self.battery_level -= 1
            if self.battery_level <= 0:
                print("배터리가 모두 소모되었습니다. 로봇이 멈춥니다.")
                self.stop_cleaning()
                self.is_stuck = True # 배터리 방전도 낑김 상태로 간주 (예시)
                self.display_status() # 상태 표시
        elif not self.is_cleaning:
            print("청소 중이 아니므로 이동하지 않습니다.")
        elif self.is_stuck:
            print("로봇이 낑겨 있어 이동할 수 없습니다.")

    def detect_obstacle(self):
        # 장애물 감지 로직 (임시)
        if self.is_cleaning and not self.is_stuck:
            obstacle_detected = input("장애물을 감지했습니까? (y/n): ").lower()
            if obstacle_detected == 'y':
                print("장애물을 감지했습니다. 경로를 변경합니다.")
                # 경로 변경 로직 (예: 다른 방향으로 이동)
                self.move("뒤") # 예시로 뒤로 이동
                return True
            else:
                return False
        return False

    def detect_dirty_spot(self):
        # 더러운 곳 감지 로직 (임시)
        if self.is_cleaning and not self.is_stuck:
            dirty_detected = input("더러운 곳을 감지했습니까? (y/n): ").lower()
            if dirty_detected == 'y':
                print("더러운 곳을 감지했습니다. 집중 청소를 시작합니다.")
                # 집중 청소 로직 (예: 해당 구역 반복 청소)
                print("집중 청소 중...")
                self.battery_level -= 5 # 집중 청소 시 배터리 더 소모
                if self.battery_level <= 0:
                    print("배터리가 모두 소모되었습니다. 로봇이 멈춥니다.")
                    self.stop_cleaning()
                    self.is_stuck = True
                    self.display_status()
                return True
            else:
                return False
        return False

    def move_to_charging_station(self):
        if not self.is_charging and not self.is_stuck:
            print("충전 스테이션으로 이동합니다.")
            # 충전 스테이션으로 이동하는 복잡한 로직은 생략하고 단순하게 표현
            print("... 충전 스테이션 도착 ...")
            self.start_charging()
        elif self.is_charging:
            print("이미 충전 중입니다.")
        elif self.is_stuck:
            print("로봇이 낑겨 있어 충전 스테이션으로 이동할 수 없습니다.")

    def start_charging(self):
        if not self.is_charging and not self.is_stuck:
            print("충전을 시작합니다.")
            self.is_charging = True
            self.is_cleaning = False # 충전 중에는 청소 중단
            # 충전 로직 (시간이 지남에 따라 배터리 증가)
            # 실제로는 타이머나 쓰레드를 사용하여 충전 상태를 관리
            print("... 충전 중 ...")
            self.battery_level = 100 # 단순화를 위해 즉시 완충
            print("충전이 완료되었습니다.")
            self.stop_charging()
        elif self.is_charging:
            print("이미 충전 중입니다.")
        elif self.is_stuck:
            print("로봇이 낑겨 있어 충전할 수 없습니다.")


    def stop_charging(self):
        if self.is_charging:
            print("충전을 중단합니다.")
            self.is_charging = False
        else:
            print("현재 충전 중이 아닙니다.")

    def handle_stuck(self):
        if self.is_stuck:
            print("로봇이 낑겼습니다. 사용자의 도움이 필요합니다.")
            # 실제 로봇에서는 알림이나 다른 방식으로 사용자에게 알림
            release = input("로봇을 낑김 상태에서 해제했습니까? (y/n): ").lower()
            if release == 'y':
                self.is_stuck = False
                print("로봇이 낑김 상태에서 해제되었습니다. 작업을 재개합니다.")
                if not self.is_charging and self.battery_level > 20:
                    self.start_cleaning() # 청소 중이었다면 다시 시작
                elif self.battery_level <= 20:
                     self.move_to_charging_station() # 배터리가 부족하면 충전소로
            else:
                print("낑김 상태가 지속됩니다.")
        else:
            print("로봇은 낑겨 있지 않습니다.")

# 로봇 청소기 인스턴스 생성
my_robot = RobotVacuum()

# 로봇 청소기 시뮬레이션 실행
while True:
    my_robot.display_status()
    action = input("원하는 동작을 입력하세요 (청소시작, 청소중단, 이동[앞/뒤/왼쪽/오른쪽], 충전, 낑김처리, 종료): ").lower()

    if action == "청소시작":
        my_robot.start_cleaning()
    elif action == "청소중단":
        my_robot.stop_cleaning()
    elif action.startswith("이동"):
        parts = action.split()
        if len(parts) == 2:
            direction = parts[1]
            if direction in ["앞", "뒤", "왼쪽", "오른쪽"]:
                my_robot.move(direction)
                my_robot.detect_obstacle() # 이동 후 장애물 감지 시뮬레이션
                my_robot.detect_dirty_spot() # 이동 후 더러운 곳 감지 시뮬레이션
            else:
                print("유효하지 않은 이동 방향입니다.")
        else:
            print("이동 명령어가 올바르지 않습니다. 예: '이동 앞'")
    elif action == "충전":
        my_robot.move_to_charging_station() # 충전 스테이션으로 이동 후 충전 시작
    elif action == "낑김처리":
        my_robot.handle_stuck()
    elif action == "종료":
        print("로봇 청소기 시뮬레이션을 종료합니다.")
        break
    else:
        print("알 수 없는 명령어입니다.")

```
